## Question #4 
You are managing a list of students enrolled in two different courses. You need to perform various set operations to understand the student enrollment. Task: Create two sets: one for students in "Course A" and one for students in "Course B". Find the students who are in "Course A" but not in "Course B" (difference). Find students who are only in one of the two courses (symmetric difference). course_a = {"John", "Alice", "Bob", "David"} course_b = {"Alice", "Eve", "Charlie", "David"}

Using set methods

• Find students who are in Course A but not in Course B (difference)

• Find students who are only in one of the two courses (symmetric difference)

```python
course_a = {"John", "Alice", "Bob", "David"}
course_b = {"Alice", "Eve", "Charlie", "David"}

difference = course_a.difference(course_b)
symmetric_difference = course_a.symmetric_difference(course_b)
print("Students in Course A but not in Course B:", difference)
print("Students only in one of the two courses:", symmetric_difference)
```

## Question #7
inventory = { &quot;The Great Gatsby&quot;: {&quot;price&quot;: 10.99, &quot;quantity&quot;: 5}, &quot;Rich dad poor dad&quot;: {&quot;price&quot;:
8.99, &quot;quantity&quot;: 8}, &quot;Harry Potter&quot;: {&quot;price&quot;: 12.50, &quot;quantity&quot;: 2}, }
• Add a new book in inventory using method and get all book titles using method
• Remove a book if it&#39;s out of stock

```python
inventory = {
    "The Great Gatsby": {"price": 10.99, "quantity": 5},
    "Rich dad poor dad": {"price": 8.99, "quantity": 8},
    "Harry Potter": {"price": 12.50, "quantity": 2}
}
inventory.update({
    "Deep Work": {"price": 11.99, "quantity": 0}
})

book_titles = list(inventory.keys())
print("Book Titles:", book_titles)

for book in list(inventory.keys()):
    if inventory[book]["quantity"] == 0:
        inventory.pop(book)

print("Updated Inventory:", inventory)
```

## Question #10 
Set Comparison
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7}

Find common elements.
Find elements only in Set A.
Find elements only in Set B.

```python
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7}

Common_elements = A.intersection(B)
only_A = A - B
only_B = B - A

print("Common elements:", Common_elements)
print("Only in Set A:", only_A)
print("Only in Set B:", only_B)
```
