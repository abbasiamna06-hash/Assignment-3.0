## Question 1:

Write a Python program that asks the user to guess a secret word.

Keep asking until the user enters the correct word ("python").

When guessed correctly, print "You guessed it!"

```python
secret_word = "python"

guess_word = input("Guess the word: ")

while guess_word != secret_word:
    guess_word = input("Guess the word: ")

print("You guessed it!")
```

## Question 2

Stop sending emails after 3 users (Using for loop and break)

users = ["Ali", "Sara", "John", "Zara", "Mike"]

```python
users = ["Ali", "Sara", "John", "Zara", "Mike"]

count = 0

for user in users:
    print("Email sent to", user)
    count += 1

    if count == 3:
        print("Limit reached. Stop sending emails.")
        break
```

## Question 3

Stop monitoring temperature if critical temperature reached (Using for loop and break)

temps = [22, 25, 30, 45, 28] critical = 40

```python
temps = [22, 25, 30, 45, 28]
critical = 40

for temp in temps:
    if temp >= critical:
        print("Critical temperature reached!")
        break
    print("Temperature:", temp)
```

## Question 4. 

Calculate total price of items in a shopping cart (Using For loop)

cart = [100, 250, 50, 300]

```python
cart = [100, 250, 50, 300]

total = 0

for price in cart:
    total += price

print("Total Price:", total)    
```

## Question 5. 

Skip customers with unpaid bills (Using For loop and continue)

customers = [ {"name": "Ali", "paid": True},

{"name": "Sara", "paid": False},

{"name": "John", "paid": True}
]

```python
customers = [
    {"name": "Ali", "paid": True},
    {"name": "Sara", "paid": False},
    {"name": "John", "paid": True}
]

for customer in customers:
    if customer["paid"] == False:
        continue

    print("Access granted to:", customer["name"])
```

## Question 7. 

Extract domain names from email list (Using for loop)

emails = ["ali@gmail.com", "sara@yahoo.com", "john@hotmail.com"]

```python
emails = ["ali@gmail.com", "sara@yahoo.com", "john@hotmail.com"]

domains = []

for email in emails:
    domain = email.split("@")[1]
    domains.append(domain)

print("domains:", domains)
```

## Question 8. 

Print only overdue tasks (Using for loop)

tasks = [ {"task": "Pay bills", "status": "done"},

{"task": "Submit report", "status": "pending"},

{"task": "Clean house", "status": "pending"}
]

```python
tasks = [
    {"task": "Pay bills", "status": "done"},
    {"task": "Submit report", "status": "pending"},
    {"task": "Clean house", "status": "pending"}
]

for task in tasks:
    if task["status"] == "pending":
        print("Overdue:", task["task"])
```

## Question 9:

Write a Python program that keeps asking the user to enter numbers and prints them.

The program should stop the loop if the user enters a negative number.

Use a while True loop with a break statement

```python
while True:
    number = int(input("Enter a number (negative to stop): "))

    if number < 0:
        print("Negative number entered. Loop stopped!")
        break

    print("You entered:", number)
```

## Question 10. 

Skip weekends while printing workdays (Using for loop and continue)

days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

```python
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

for day in days:
    if day == "Saturday" or day == "Sunday":
        continue

    print("Workday:", day)
```
